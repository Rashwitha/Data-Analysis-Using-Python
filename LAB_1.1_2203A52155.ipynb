```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as pyplt
```

```python
#a) Create vectors, factors, lists, matrices, and data frames.
insurance_data = pd.read_csv("/content/insurance.csv")
print(insurance_data.head())
```

```python
#vector
age_vector = insurance_data['age']
print("\nVector (Age):")
print(age_vector)
```

```python
#factor
sex_factor=insurance_data['sex']
print("\n factor(sex):")
print(sex_factor)
```

```python
#list
insurance_list = [age_vector.tolist(), sex_factor.tolist(), insurance_data]
print("\nList (Age, Sex):")
print(insurance_list)
```

```python
# matrix
insurance_matrix = insurance_data[['age', 'children']].to_numpy()
print("\nMatrix (Age and Children):")
print(insurance_matrix)
```

```python
# data frame
insurance_dataframe = insurance_data[['age', 'smoker']]
print("\nData Frame (Age and Smoker):")
print(insurance_dataframe)
```

```python
#b) Generate a scatter plot. Customize it by adding a title, axis labels, text annotations, and changing the plot characters (e.g., circles to plus signs).
import pandas as pd
import matplotlib.pyplot as plt
x = insurance_data['age']
y = insurance_data['bmi']

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='green', marker='+', s=100, label='Age vs BMI')


plt.title("Scatter Plot of age vs bmi", fontsize=16, fontweight='bold')
plt.xlabel("age", fontsize=12)
plt.ylabel("bmi", fontsize=12)

for i in range(10):
    plt.text(x.iloc[i] + 1, y.iloc[i], f"({x.iloc[i]}, {y.iloc[i]:.1f})", fontsize=9)

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend()
plt.show()
```

```python
# Scatter plot Age vs Smoking
plt.figure(figsize=(8, 6))
sns.scatterplot(data=insurance_data, x='age', y='smoker', marker='+', color='green')
plt.title("Scatter Plot: Age vs Smoking")
plt.xlabel("Age")
plt.ylabel("Smoker (Yes/No)")
plt.grid(alpha=0.5)
plt.show()
```

```python
#Plot a scatter plot between age(20 to 40) and smoking
#Age between 20 and 40
filtered_data = insurance_data[(insurance_data['age'] >= 20) & (insurance_data['age'] <= 40)]

# Scatter plot: Age (20-40) vs Smoking
plt.figure(figsize=(8, 6))
sns.scatterplot(data=filtered_data, x='age', y='smoker', marker='+', color='blue')
plt.title("Scatter Plot: Age (20-40) vs Smoking")
plt.xlabel("Age")
plt.ylabel("Smoker (Yes/No)")
plt.grid(alpha=0.5)
plt.show()
```

```python
#Find expenses of smokers and non smokers,(plot)
smoker_expenses = insurance_data.groupby('smoker')['expenses'].sum()
print(smoker_expenses)
```

```python
#Find the number of smokers gender wise (plot)
smoker_gender_count = insurance_data[insurance_data['smoker'] == 'yes']['sex'].value_counts()
print(smoker_gender_count)
```

```python
#Plot a scatter plot between age and expenses
plt.figure(figsize=(8, 6))
sns.scatterplot(data=filtered_data, x='age', y='expenses', marker='+', color='grey')
plt.title("Scatter Plot: age and expenses")
plt.xlabel("Age")
plt.ylabel("Expenses")
plt.grid(alpha=0.5)
plt.show()
```

```python
#count the teenage smokers
teenage_smokers = insurance_data[(insurance_data['age'] >= 13) & (insurance_data['age'] <= 19) & (insurance_data['smoker'] == 'yes')]
teenage_smokers_count = teenage_smokers.shape[0]
print(teenage_smokers_count)
```
